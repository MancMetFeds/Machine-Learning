```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
path = './data/melb_data.csv'
melbData = pd.read_csv(path)
melbData
```

```python
melbData.describe()
```

### First ML Model

```python
melbData.columns
```

```python
# Drop the missing data with missing information
melbData = melbData.dropna(axis=0)
```

```python
response_y = melbData.Price 
```

```python
model_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude', 'Longtitude']
```

```python
X = melbData[model_features]
```

```python
X.describe()
# Shows the top few rows, 
# This is a useful step as it may highlight details
# that may require further investigation
X.head()
```

```python
melb_train_X, melb_val_X, melb_train_y, melb_val_y = train_test_split(X, response_y, random_state = 1)


melb_model = DecisionTreeRegressor(random_state = 1 ) # random state : data seed

melb_model.fit(melb_train_X, melb_train_y)
```

### Predicting prices for houses entering the market 

```python
print(X.head())
# uses the model to predict prices for the first few houses in the feature set
melb_pred = melb_model.predict(melb_val_X)
print(melb_pred)
```

## Model Validation

```python
meanAbsErr = mean_absolute_error(melb_val_y, melb_pred)
meanAbsErr
```

```python
# Loading Data
path = './data/train.csv'
iowaData = pd.read_csv(path)
```

```python
# Summary Statistics
iowaData.describe()
```

```python
avgLotSize = round(iowaData.LotArea.mean())
avgLotSize
```

```python
from datetime import datetime
newestHomeAge = datetime.now().year - iowaData.YearBuilt.max()
newestHomeAge
```

```python
iowaData.columns
```

```python
response_var = iowaData.SalePrice 
```

```python
# According to Kaggle: no NA data so no need for DataFrame.dropna()
# iowaData.dropna(axis=0)
model_feature_set = ["LotArea","YearBuilt","1stFlrSF","2ndFlrSF","FullBath","BedroomAbvGr","TotRmsAbvGrd"]
iowa_X = iowaData[model_feature_set]
iowa_X.head()
```

```python
train_X, test_X, train_response_var, test_response_var = train_test_split(iowa_X, response_var, random_state = 0)
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
best_tree_size = 0
lowest_mae = 0

for n in [5, 25, 50, 100, 250, 500]:
    mae = get_mae(n, train_X, test_X, train_response_var, test_response_var)
    if lowest_mae == 0:
        lowest_mae = mae
    elif mae < lowest_mae:
        best_tree_size = n
        lowest_mae = mae 
    #print("node count: %d\t\tm.a.e : %d" %(n, mae))
print("the lowest mean absolute error is %d for %d nodes" %(lowest_mae, best_tree_size))
```

```python
iowa_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state = 0)

iowa_model.fit(train_X, train_response_var)
```

```python
pred = iowa_model.predict(test_X)
pred

meanAbsErr = mean_absolute_error(test_response_var, pred)
meanAbsErr
```
